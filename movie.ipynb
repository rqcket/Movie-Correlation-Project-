```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #Adjusts the configuration of the plots we will create 


#Read in the data

df= pd.read_csv("movies.csv")
```

```python
#Looking at the data

df.head(30)
```

```python
#Cleaning the data

#Checking for empty columns 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
df=df.dropna()
```

```python
df.dtypes
```

```python
df
```

```python
#Changing column data types

df['budget']=df['budget'].astype('int64')

df['gross']=df['gross'].astype('int64')

df['votes']=df['votes'].astype('int64')

df['runtime']=df['runtime'].astype('int64')
```

```python
#Creating an year column

df['Year'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#Dropping Duplicates

df.drop_duplicates()
```

```python
#Checking for correlation between budget and gross 
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Movie')

plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#Things which correlate the most with Gross Revenue

df.corr(method='pearson')
```

```python
correlation_matrix= df.corr(method='pearson')


sns.heatmap(correlation_matrix,annot=True)


plt.title('Correlation Matrix for Numeric Features of the Data')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
